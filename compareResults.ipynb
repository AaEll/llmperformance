```python
import json
with open('run_results_llama_70bb.json', 'r') as f:
    json_70b = json.loads(f.read())
with open('run_results_falcon_30b.json', 'r') as f:
    json_7b = json.loads(f.read())
with open('run_results_flan_540b.json', 'r') as f:
    flan_35b = json.loads(f.read())
```

```python

overall = {
    'same_pred' : 0,
    'correct_pred_7b' : 0,
    'correct_pred_70b' : 0,
    'union_correct' : 0,
    'same_answer' : 0,
    'total' : 0,
}

for k in json_7b.keys():
    print(k)
    same_pred = 0
    correct_pred_7b = 0
    correct_pred_70b = 0
    union_correct = 0
    same_answer = 0
    answer_set = set([])

    total = 0

    for pred_70b, gold_70b, pred_35b, gold35b, pred_7b, gold_7b in zip(json_70b[k]['pred_answers'],json_70b[k]['gold_answers'],flan_35b[k]['pred_answers'],flan_35b[k]['gold_answers'],json_7b[k]['pred_answers'],json_7b[k]['gold_answers']):
        
        if (pred_70b == gold_70b):
            correct_pred_70b+=1
        if (pred_7b == gold_7b):
            correct_pred_7b+=1
        if (pred_7b == pred_70b):
            same_pred+=1
        if (gold_7b == gold_70b):
            same_answer+=1
        if (pred_35b == gold35b) or (pred_7b == gold_7b) :
            #if (pred_7b == gold_7b) or (pred_70b == gold_70b) or (pred_35b == gold35b):
            union_correct+=1
        total+=1

        answer_set.add(pred_7b)
        answer_set.add(pred_35b)

    print(union_correct/total)
    print(same_pred/total)
    print(same_answer/total)
    print(correct_pred_70b/total)
    print(correct_pred_7b/total)

    overall['same_answer'] += same_answer
    overall['correct_pred_7b'] += correct_pred_7b
    overall['correct_pred_70b'] += correct_pred_70b
    overall['union_correct'] += union_correct
    overall['same_pred'] += same_pred
    overall['total'] += total

print(overall)
```
